# Graphs

Additional notebook with graphical analysis.

```python
# Read the data
import pandas as pd
ped = pd.read_csv("data/final/ped_filtered.csv")
```

```python
ped.head()
```

```python
# The best seected features
features_names_top = ['exm-rs12913832', 'rs1129038', 'rs2238289']
```

```python
import matplotlib.pyplot as plt
```

```python
for poly in list(features_names_top):
    print(poly)
    (ped
     .groupby(['color', poly])
     .size()
     .unstack()
     .plot.bar(stacked=True)
    )
    plt.show()
```

```python
for poly in list(ped.columns.values):
    print(poly)
    (ped
     .groupby(['color', poly])
     .size()
     .unstack()
     .plot.bar(stacked=True)
    )
    plt.show()
```
