# Detect useless polymorfphisms

In this notebook the oryginal dataset is analyzed and we detect (and remove from further analysis) polymorphisms that has no ability to differentiate between eyes colors.

### Transform source files to csv

```python
PED_FILE = "data/source/HACK_15_one.ped"  # Data
MAP_FILE = "data/source/HACK_15_one.map"  # Headers
FINAL_FILE = "data/final/ped.csv"  # Resulting csv

from ped_to_csv import ped_to_csv
ped_to_csv(PED_FILE, MAP_FILE, FINAL_FILE)
```

### First analysis

```python
import pandas as pd
```

```python
ped = pd.read_csv(FINAL_FILE)
```

```python
ped.head()
```

```python
ped.groupby('color').count()
```

### Data clearing

```python
# Remove unknown values
ped = ped.drop(ped[ped.color == -9].index)
```

```python
ped.groupby('color').count()
```

```python
# Remove not used columns
ped = ped.drop(['id', 'fid', 'cos1', 'cos2', 'sex'], axis=1)
```

### First useless detection

Here we identify and remove form further analysis polymorphisms that in fact doesn't exists in our data
(one and the same variant for both eyes colors (classes))

```python
# List and remove columns with only one value
useless_poly = []
for poly in list(ped.columns.values):
    if ped[poly].nunique() == 1:
        print(poly)
        useless_poly.append(poly)
        ped = ped.drop(poly, axis=1)
```

```python
# Shape after first clearing
ped.shape
```

### Second useless detection

Here we identify and remove form further analysis polymorphisms that in fact doesn't exists in our data
(one and the same variant for both eyes colors (classes))

```python
#Remove columns with two values where one of them is (0,0)
for poly in list(ped.columns.values):
    if ped[poly].nunique() == 2:
        if "('0', '0')" in ped[poly].unique():
            print(poly, ped[poly].unique())
            useless_poly.append(poly)
            ped = ped.drop(poly, axis=1)
```

```python
# Shape after second clearing
ped.shape
```

```python
# Save useless to file
with open("useless_polymorphisms.txt", "w") as f:
    for poly in useless_poly:
        f.write(poly+"\n")
```

```python
# Store data without useless polymorphisms in 'ped_filtered.csv' file
ped.to_csv("data/finals/ped_filtered.csv")
```
